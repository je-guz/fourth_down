```python
# Function that determines switch in possession **by row val on sorted dataframe
# list_of_possession - list grabbed from column of dataframe "OffenseTeam"
# @returns list of row indices in dataframe of possession changes (value inclusive)


def possession_values(list_of_possession):
    change_list =[]
    team1 = list_of_possession[0]
    for i in range(len(list_of_possession)):
        if (list_of_possession[i] != team1):
            change_list.append(i)
            team1 = list_of_possession[i]
            
    return change_list
        

    
```

Create a function that will help calculate expected points on atempted fourth down decision (ie field goal, 1st down conversion)

```python
#Function that keeps track of field goal attempts/completions & conversion attempts/completions And points for each (in a game)
#game_dataframe - clean dataframe of any given game
#possession list - list of row indices in dataframe of possession changes (value inclusive)
#team - team we are looking at 
def count(game_dataframe, possession_list, team):
    #counters
    fg_ep = 0
    fg_att = 0
    fg_s = 0
    
    conversion_ep = 0
    conversion_att = 0
    conversion_s = 0
    
    
    #iterate over each drive
    for i in range(len(possession_list)):
        

        #if last drive
        if (i == len(possession_list)-1):
            
            drive = game_dataframe[possession_list[i]:]
            
        #if any other drive    
        else:
            drive = game_dataframe[possession_list[i]:possession_list[i+1]]

        if (drive.iloc[0].get("OffenseTeam") != team):
            continue
        #if any fourth down plays    
        if (drive[drive.get("Down") == 4].count()[0] > 0):
            
            #from first fourth play 
            first_fourth = drive[drive.get("Down") == 4].index[0]
            #plays after first fourth down
            plays_after_fourth = drive.loc[first_fourth:]

            
            #for every play after fourth, check for result of drive
            for j in range(len(plays_after_fourth)):
                   
                play = plays_after_fourth.iloc[j]
                #if attempted field goal
                if (play.get("PlayType") == ("FIELD GOAL")):

                   
                    fg_att += 1
                   
                    #if field goal successful 
                    if ("NO GOOD" not in play.get("Description")):
                        #if field goal comes after successful 4th down conversion, average attempts/point
                        if (j > 0):
                            conversion_s+=1
                            
                            conversion_att -= 1
                            conversion_ep += 1.5
                            fg_ep+= 1.5
                        else:
                            fg_s += 1
                            fg_ep+= 3
                 
                elif (play.get("PlayType") == ("PUNT")):
                    if (j > 1):
                        conversion_s+=1
                        

                else:
                    #attempt conversion to 1st down
                    if (play.get("Down") == 4):
                        conversion_att += 1
                        #best case scenario - there is a score on the following plays (as a result of conversion)
                        if plays_after_fourth.get("IsTouchdown").sum() > 0:
                            conversion_s+=1
                            conversion_ep += 7
                            break
                
    return [fg_att, fg_ep, conversion_att, conversion_ep, fg_s, conversion_s]
        
            
    
```

Run counts of field goal attempts/ points & fourth down conversion attempts/points for a season

```python
#Function that runs above function for all games in a season 
def over_season(season_df, unique_games):
    fg_ep = 0
    fg_att = 0
    fg_s = 0
    
    conversion_ep = 0
    conversion_att = 0
    conversion_s = 0
    
    for i in range(len(unique_games)):
        ith_game = season_df[season_df.get("GameId") == unique_games[i]].sort_values(by = ["Quarter",  "Minute", "Second"], ascending = (True, False, False))
        
        possession_list = possession_values(ith_game.get("OffenseTeam").tolist())
        teams = [ith_game.get("OffenseTeam").iloc[0], ith_game.get("DefenseTeam").iloc[0]]
        for j in range(len(teams)):
            counts = count(ith_game, possession_list, teams[j])
            fg_att += counts[0]
            fg_ep += counts[1]
            conversion_att += counts[2]
            conversion_ep += counts[3]
            fg_s+= counts[4]
            conversion_s += counts[5]
        if (fg_att != 0):
            fg_rate = fg_ep/fg_att
            f = fg_s/fg_att
        else:
            fg_rate = 0
            f = 0
        if (conversion_att != 0):
            conversion_rate = conversion_ep/conversion_att
            c = conversion_s/conversion_att
        else:
            conversion_rate = 0
            c = 0
    return [fg_rate, conversion_rate, f, c]
        
```

```python
#Create function to seperate games with fourth down drives within 3 bins [=<3, =<6, =<9] yards to go 

def ToGo_Binning(season_df, unique_games):
    low_yards_togo = season_df[(season_df.get("BinToGo") == "Low") |  (season_df.get("BinToGo") == "NA")]
    med_yards_togo = season_df[(season_df.get("BinToGo") == "Med") |  (season_df.get("BinToGo") == "NA")]
    high_yards_togo = season_df[(season_df.get("BinToGo") == "High" ) |  (season_df.get("BinToGo") == "NA")]
    long_yards_togo = season_df[(season_df.get("BinToGo") == "Long")  |  (season_df.get("BinToGo") == "NA")]
    
    bins = [low_yards_togo, med_yards_togo, high_yards_togo, long_yards_togo]
    binning_results = []
    for i in  range(len(bins)):
        binning_results.append(over_season(bins[i], unique_games))
    return binning_results
        
```

```python
#Create function to seperate games with fourth down drives with 4 bins [=<25, <=50, <=75, <=100] starting field position

def FieldPos_Binning(season_df, unique_games):
    first_section = season_df[(season_df.get("FieldbyFourth") == "FirstFourth") |  (season_df.get("FieldbyFourth") == "NA")]
    second_section = season_df[(season_df.get("FieldbyFourth") == "SecondFourth") |  (season_df.get("FieldbyFourth") == "NA")]
    third_section = season_df[(season_df.get("FieldbyFourth") == "ThirdFourth" ) |  (season_df.get("FieldbyFourth") == "NA")]
    fourth_section = season_df[(season_df.get("FieldbyFourth") == "LastFourth")  |  (season_df.get("FieldbyFourth") == "NA")]
    
    bins = [first_section, second_section, third_section, fourth_section]
    binning_results = []
    for i in  range(len(bins)):
        binning_results.append(over_season(bins[i], unique_games))
    return binning_results
        
```

```python
def epa(season_df, yards_to_go, field_pos):
    if (yards_to_go <= 3):
        df = season_df[(season_df.get("BinToGo") == "Low") |  (season_df.get("BinToGo") == "NA")]     
    elif (yards_to_go > 3) & (yards_to_go <= 6):
        df = season_df[(season_df.get("BinToGo") == "Med") |  (season_df.get("BinToGo") == "NA")]
    elif (yards_to_go > 6) & (yards_to_go <= 10):
        df = season_df[(season_df.get("BinToGo") == "High" ) |  (season_df.get("BinToGo") == "NA")]
    elif (data_2020['ToGo'] > 10):
        df = season_df[(season_df.get("BinToGo") == "Long")  |  (season_df.get("BinToGo") == "NA")]
    df = df[((df['Down'] == 4) & (df['YardLine'] > field_pos - 5)) & 
                  ((df['Down'] == 4) & (df['YardLine'] <= field_pos + 5)) | (df['Down'] != 4) ]  
    unique = df.get("GameId").unique()
    
    return over_season(df, unique )
    
```

```python
#match dates in different formats -> ex input (1/3/2021, 2021-01-03)
def ToSlash(dash_date):
    date = dash_date.split("-")
    toreturn = str(int(date[1])) + "/" + str(int(date[2])) + "/" + str(int(date[0]))
    return toreturn
    
    
    
```

```python
#Match games with spread & score differential 
def add_to_df(season_df, final_spread_df):
    spread = {}
    final_score = {}
    games = season_df["GameId"].unique()
    for game in games:
        
        
        game_df = season_df[season_df.get('GameId')== game].iloc[0]
    
        date = game_df.get("GameDate")
        offense = game_df.get("OffenseTeam")
        defense = game_df.get("DefenseTeam")
        

        
        date_converted = ToSlash(date)
        
        gamesInDate = final_spread_df[final_spread_df.get("schedule_date") == date_converted]
        
        for i in range(len(gamesInDate)):
        
            currgame = gamesInDate.iloc[i]
            possTeam = currgame.get('team_favorite_id')
            
            if ((offense == possTeam) | (defense == possTeam)):
                spread[game] = [possTeam, currgame.get("spread_favorite")]
                final_score[game] = [currgame.get("team_home"), currgame.get("score_home"), currgame.get("score_away")]
                
                break
     
    return [spread, final_score]
    
        
    
```

```python
#Avg change in points over next x second 

#over game df
#look at next x seconds of playtime
# find expected change in score diff  
def scoring(game_df, quarter, minutes):
    
    g = game_df[game_df.get("Quarter") >= quarter]
    g2 = g[g.get("Minute") <= minutes]

    teams = [game_df.get("OffenseTeam").iloc[0], game_df.get("DefenseTeam").iloc[0]]
    team1_p = 0
    team2_p = 0
    #plays after first fourth down

            
    #for every play after fourth, check for result of drive
    for j in range(len(g2)):
                   
        play = g2.iloc[j]
        #if attempted field goal
        if (play.get("PlayType") == ("FIELD GOAL")): 

            #if field goal successful 
            if ("NO GOOD" not in play.get("Description")):
                if (play.get("OffenseTeam") == teams[0]):
                    team1_p += 3
                    print(play.get("Description"))
                else:
                    team2_p += 3
                    print(play.get("Description"))
                    
                 
        elif (play.get("PlayType") == ("TWO-POINT CONVERSION")):
            if ("SUCCEEDS" in play.get("Description")):
                if (play.get("OffenseTeam") == teams[0]):
                        team1_p += 2
                        print(play.get("Description"))
                else:
                    team2_p += 2
                    print(play.get("Description"))
        elif (play.get("PlayType") == ("EXTRA POINT")):
            if ("NO GOOD" not in play.get("Description")):
                if (play.get("OffenseTeam") == teams[0]):
                    team1_p += 1
                    print(play.get("Description"))
                else:
                    team2_p += 1
                    print(play.get("Description"))

        elif (play.get("IsTouchdown") == 1):
            if (play.get("OffenseTeam") == teams[0]):
                    team1_p += 6
                    print(play.get("Description"))
            else:
                team2_p += 6
                print(play.get("Description"))
        if (max(team1_p, team2_p) == team1_p):
            winning = teams[0]
        else:
            winning = teams[1]
    return [winning, abs(team1_p-team2_p)]
    
```

```python
def time_diff(game_df):
    normal_df = game_df[game_df.get("TimeSection") == 'Other']
    q2_df = game_df[game_df.get("TimeSection") == 'EndSecondQ']
    q4_df = game_df[game_df.get("TimeSection") == 'EndFourthQ']
    
    return [(45*60)/len(normal_df), (10*60)/len(q2_df), (5*60)/len(q4_df)]
```

```python
def def_weights(season_df):
    unique = season_df.get("GameId").unique()
    weights = [0,0,0]
    for i in range(len(unique)):
        ith_game = season_df[season_df.get("GameId") == unique[i]].sort_values(by = ["Quarter",  "Minute", "Second"], ascending = (True, False, False))
        w = time_diff(ith_game)
        weights[0] += w[0]
        weights[1] += w[1]
        weights[2] += w[2]
        final_weights = [weights[0]/len(unique), weights[1]/len(unique), weights[2]/len(unique)]
    print(final_weights)
    return [final_weights[0]/final_weights[1], final_weights[0]/final_weights[2]]
        
    
```
